# Number of every country's male and female atheletes

```python
country_gender_count=df.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
# Number of gold medals won by Country
country_medal_won=df[df["Medal"]=="Gold"].groupby("NOC")["Medal"].count()
print(country_medal_won)
```

```python
country_medal_won[country_medal_won==747]
```

```python
Sports_gender_avg_weight=df.groupby(["Sport","Sex"])["Weight"].mean()
print(Sports_gender_avg_weight)
```

```python
sport_event_count=df.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count.plot(kind="bar")
plt.title("Number of unique event per sport")
plt.xlabel("Sport")
plt.ylabel("Number of Unique Event")
plt.xticks(rotation=70)
plt.show()
```

```python
year_participant_count=df.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="bar")
plt.title("No of Participant per year")
plt.xlabel("Year")
plt.ylabel("No of Participants")
plt.show()
```

```python
country_avg_age=df.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.head(10).plot(kind="bar",color="Red")
plt.title("Top 10 countries with highest age of participants")
plt.xlabel("Country")
plt.ylabel("Avg Age")
plt.xticks(rotation=45)
plt.show()
```

```python
country_avg_age=df.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.tail(10).plot(kind="bar")
plt.title("Top 10 countries with lowest age of participants")
plt.xlabel("Country")
plt.ylabel("Avg Age")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df,x="Season",y="Age")
plt.title("distribution of Ages by Season")
```

```python
sns.violinplot(data=df,x="Medal",y="Height",palette="Set2")
plt.title("Distribution of Height By Medal")
```

```python
sns.scatterplot(data=df,x="Height",y="Weight",hue="Medal",palette="Set1")
plt.title("Athlete Height vs Weight By Medal status")
```

```python
Medals_By_country_Year=df.pivot_table(index="NOC",columns="Year",values="Medal",aggfunc="count")
sns.heatmap(Medals_By_country_Year,linewidths=0.5)
plt.title("Medals Count By Country and Year")
```
